# Function

## structure function

함수의 선언은 아래의 형식으로 한다.

- type name(arg1, arg2 ..){}

## type of function

함수는 크게 두종류로 나뉘는데, 리턴값이 없는 함수와 있는 함수이다.

리턴이 없는 함수는 void type으로, 리턴이 존재하는 함수는 리턴값의 타입으로 선언한다.
