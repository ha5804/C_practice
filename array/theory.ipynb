# understanding pointer and array together

## 1.key theory

array의 이름(배열 선언시 사용한 이름)에는 array의 첫번째 요소의 주소값이 포함되어있다.

즉 다음과 같은 ptr 선언이 가능하다.

int arr[2] = {1,2,3}

int *ptr = arr;

*ptr에 arr의 첫번째 주소저장.
즉 int *ptr = &arr[0]과 완전히 같은 의미이다.

# understandin Pointers and Functions

따라서 &val 을 사용해야한다.
