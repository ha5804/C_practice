# Relationship multi_array and pointer

## 1.Pointer types of 1d and 2d array names

- C에서 배열 이름은 포인터처럼 행동한다.
따라서, 만약 int arr[10] 이라고 선언한다면, arr은 10개의 int를 가진 배열이지만, 함수에 전달할 때는 자동으로 arr -> arr[0]을 가리키는 포인터로 변하게 된다. 따라서 인자를 *ptr로 받는 함수에 전달이 가능하다.

- 만약 int *parr[20] 이라고 선언하였다면, 이는 pointer를 20개 요소로 갖는 포인터 배열이다. 이때 parr을 함수에 전달하게 된다면, 첫 요소를 전달하기 때문에 doble pointer로 바뀌어 int **로 바뀌게 된다. 따라서 인자를 *ptr로 받는 함수에는 맞지 않게된다. 

### Solution

- 그렇다면 포인터 배열을 함수에 인자로 사용하기 위해서는, 실제 포인터 배열에 들어갈 값을 선언한후에, 실제로 그 값의 메모리 주소를 포인터 배열에 저장하여 사용하는 것이다.

## 2. 2D array & pointer

#### 2D sturcture

만약 2d array를 int arr2D[3][4]; 로 선언했다 하자.
이는 arr2D[i] -> int 4개 (i = 0, 1, 2) 구조이다.
즉, 3개의 1차원 배열 (int[4])를 묶어놓은 배열이다.
이 말은 2D 배열의 각 행은 배열 이라는것이다.

#### 함수 전달 타입

그렇다면 arr2D배열은 전달될때 한줄(행)을 가리키는 포인터 arr2D[0]를 가리키는 포인터가 된다. 즉 예시의 int 4개 요소를 갖는 형태로 전달이 되기때문에, int (*ptr)[4] 형식으로 전달이 된다.
이때 주의점은 arr2D[0] != int *ptr 이지만, arr2D[0] = int[4] 라는 것이다.
